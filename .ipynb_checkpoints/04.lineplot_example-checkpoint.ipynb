```python
import seaborn as sns
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
## may data query
may_flights = flights.query("month == 'May'")
may_flights
```

```python
## x axis == year
## y axis == passengers
sns.lineplot(data=may_flights, x='year', y='passengers')
```

```python
## x axis year, y axis passengers, data label is month 형태로 데이터 프레임을 만들어준다.
flights_wide = flights.pivot("year", "month", "passengers")
flights_wide.head()
```

```python
sns.lineplot(data=flights_wide["May"])
```

```python
# col = month (그룹), row = data x axis label, passengers is data
sns.lineplot(data=flights_wide)
```

```python
# 자동으로 집계되어 평균 및 95% 신뢰 구간을 표시한다.
sns.lineplot(data=flights, x="year", y="passengers")
```

```python
## 그룹화 의미를 할동한다. hue를 통해 그룹핑을 한다.
sns.lineplot(data=flights, x="year", y="passengers", hue="month")
## sns.lineplot(data=flights, x="year", y="passengers", hue="month", style="month")
## 그룹별로 스타일을 달리할 수 있다.
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
## event를 집계해서 숫자들을 치기때문에 오차범위가 나타난다. (의미론적 그룹)
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event")
```

```python
## 두 가지 다른 그룹화 변수를 나타내려면 hue 및 style 둘다 할 당한다.
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="region", style="event")
```

```python
# marker, dashes styling
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="event"
                ,markers=True, dashes=False)
```

```python
# 오차 구간 대신 오차 막대를 표시하고 싶을 때는 err_style 옵션을 사용한다.
# (ci = 68) 옵션은 68% 신뢰 구간(표준 오차)을 플로팅한다.
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", err_style="bars", ci=68)
```
