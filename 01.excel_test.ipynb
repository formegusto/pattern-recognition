# Pandas Excel 데이터 사용하기

```python
import pandas as pd
from pandas import DataFrame
import datetime

print("Excel Read")
```

```python
xlsx = pd.read_excel('datas.xlsx', 
                    header = None, 
                    skiprows = 2,
                    engine = 'openpyxl')
```

# 20180501 ~ 20190430
- [0] 아파트 이름
- [1] 아파트 동
- [2] 아파트 호수

```python
xlsx
```

```python
datas = {};
for col in xlsx:
    if col > 6: # startcol 7
        datas[f'{xlsx[col][0]}-{xlsx[col][1]}-{xlsx[col][2]}'] = xlsx[col][3:]
```

```python
df = DataFrame(datas)
print(df['아파트1-104-1206'])
```

```python
for data in df['아파트1-104-1206']:
    print(data);
```
