# Pandas Excel 데이터 사용하기

```python
import pandas as pd
from pandas import DataFrame, Series
from datetime import datetime, timedelta

print("Excel Read")
```

```python
xlsx = pd.read_excel('datas.xlsx', 
                    header = None, 
                    skiprows = 2,
                     chunksize = 2,
                    engine = 'openpyxl')
```

# 20180501 ~ 20190430
- [0] 아파트 이름
- [1] 아파트 동
- [2] 아파트 호수

```python
xlsx
```

```python
datas = {};
for col in xlsx:
    if col > 6: # startcol 7
        datas[f'{xlsx[col][0]}-{xlsx[col][1]}-{xlsx[col][2]}'] = xlsx[col][3:]
```

```python
def datetime_range(start, end, delta):
        current = start
        while current < end:
                yield current
                current += delta
dts = [dt.strftime('%Y%m%d %H:%M') for dt in
            datetime_range(datetime(2018,5,1,0), datetime(2019,5,1,0), timedelta(minutes = 15))]
```

```python
df = DataFrame(datas)
print(df['아파트3-102-1205'][3]);
```

```python
sDatas = []

for idx in datas['아파트1-104-1206']:
    sDatas.append(idx)

dts = [dt.strftime('%Y%m%d') for dt in
            datetime_range(datetime(2018,5,1,0), datetime(2019,5,1,0), timedelta(days = 1))]
# print(dts)

sData = {}
dtIdx = 0
for idx in range(3,len(sDatas),96):
    sData[dts[dtIdx]] = sDatas[idx: idx + 96]
    dtIdx += 1
# sDf = DataFrame({'value': sData})
```
